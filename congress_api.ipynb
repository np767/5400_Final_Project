```python
import requests
import pandas as pd

params = {
    "api_key": API_KEY,
    "format": "json",
}

response = requests.get(BASE_URL, params=params)

if response.status_code == 200:
    data = response.json()

    # view first 5 members
    for member in data.get("members", [])[:5]:
        print(
            f"{member['name']} ({member['partyName']}) - {member['state']} "
            f"→ {member['url']}"
        )
```

```python
resp = requests.get(
    "https://api.congress.gov/v3/hearing/118",
    params={"api_key": API_KEY, "format": "json"}
)

data = resp.json()

# view one hearing
first_hearing = data["hearings"][0]
print(first_hearing)
print("\nDetail URL:", first_hearing["url"])
```
